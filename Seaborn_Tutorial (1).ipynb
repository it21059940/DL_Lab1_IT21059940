```python
# Pandas for managing datasets
import pandas as pd
```

```python
# Matplotlib for additional customization
from matplotlib import pyplot as plt
%matplotlib inline
# Seaborn for plotting and styling
import seaborn as sns
```

```python
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
```

```python

# Display first 5 observations
df.head()
```

Scatter Plot

```python

# Scatterplot arguments
sns.lmplot(x='Attack', y='Defense', data=df,
           fit_reg=True, # No regression line
           hue='Stage')   # Color by evolution stage
```

Box Plots

```python
# Boxplot
plt.figure(figsize=(9,6)) # Set plot dimensions
sns.boxplot(data=df)
```

```python

# Preprocess DataFrame
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)

# New boxplot using stats_df
plt.figure(figsize=(9,6)) # Set plot dimensions
sns.boxplot(data=stats_df)
```

Violin Plot

```python
# Violin plot
plt.figure(figsize=(12,8)) # Set plot dimensions
sns.violinplot(x='Type 1', y='Attack', data=df)
```

Heat Map

```python
# Calculate correlations
corr = stats_df.corr()

corr
```

```python
# Heatmap
plt.figure(figsize=(9,8))
sns.heatmap(corr)
```

```python

# Select the relevant features
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
data = df[features]

# Calculate the Pearson correlation coefficient
pearson_corr = data.corr(method='pearson')

# Calculate the Spearman rank correlation
spearman_corr = data.corr(method='spearman')

# Set up the matplotlib figure
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

# Plot the Pearson correlation heatmap
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', ax=ax[0])
ax[0].set_title('Pearson Correlation Coefficient')

# Plot the Spearman correlation heatmap
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', ax=ax[1])
ax[1].set_title('Spearman Rank Correlation')

# Display the plots
plt.tight_layout()
plt.show()
```
