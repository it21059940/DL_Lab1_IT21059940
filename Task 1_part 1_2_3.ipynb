```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
scale = 1
array_one = np.random.exponential(scale, (4, 4))

print(array_one)
```

```python
scale = 1.0

plt.hist(np.random.exponential(scale, 100000), density=True, bins=100, histtype="step", color="green", label="Exponential")
plt.hist(np.random.rand(100000), density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(np.random.randn(100000), density=True, bins=100, histtype="step", color="red", label="Normal")

plt.axis([-2.5, 8, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")

plt.show()
```

```python
#Increasing bin value -> to get detailed graph

scale = 1.0

plt.hist(np.random.exponential(scale, 100000), density=True, bins=150, histtype="step", color="green", label="Exponential")
plt.hist(np.random.rand(100000), density=True, bins=150, histtype="step", color="blue", label="Uniform")
plt.hist(np.random.randn(100000), density=True, bins=150, histtype="step", color="red", label="Normal")

plt.axis([-2.5, 8, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")

plt.show()
```

```python
#Using a different plot -> Set plot view to make visualization better

scale = 1.0

sns.kdeplot(np.random.exponential(scale, 100000), color="green", label="Exponential")
sns.kdeplot(np.random.rand(100000), color="blue", label="Uniform")
sns.kdeplot(np.random.randn(100000), color="red", label="Normal")

plt.xlim(-2.5, 8)
plt.ylim(0, 1.1)
plt.legend(loc="upper right")
plt.title("Kernel Density Estimate of Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")

plt.show()
```

```python
x = np.linspace(-5, 5)
y = np.linspace(-5, 5)

X, Y = np.meshgrid(x, y)

Z = X**2 + Y**2


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)

ax.set_title('3D Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
