```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading the dataset
df = pd.read_csv('Au_nanoParticle_dataset.csv')
df.head()
```

```python
# Creating a new dataframe with filtered columns
final_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]

# Displaying the first 20 rows
final_df.head(20)
```

```python
stats = final_df.describe().T
stats['Mean'] = final_df.mean()
stats['SD'] = final_df.std()
stats['25%'] = final_df.quantile(0.25)
stats['50%'] = final_df.quantile(0.5)
stats['75%'] = final_df.quantile(0.75)

stats = stats[['Mean', 'SD', '25%', '50%', '75%']]
print(stats)
```

```python
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

# Plotting each feature's histogram
features = ['N_total', 'N_bulk', 'N_surface', 'R_avg']
for i, feature in enumerate(features):
    axes[i].hist(final_df[feature], bins=30, color='skyblue', edgecolor='black')
    axes[i].set_title(f'Histogram of {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

sns.pairplot(final_df)
plt.show()
```

```python
g = sns.PairGrid(final_df)

g.map_upper(sns.histplot)

g.map_diag(sns.histplot, kde=True)

g.map_lower(sns.kdeplot, cmap="coolwarm")

plt.show()
```
